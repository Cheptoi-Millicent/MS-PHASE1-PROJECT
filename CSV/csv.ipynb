Creating CSV - Student Records


|student 1|student 2|student3|
|----------|----------|--------|
|71        |62        |78      |
|65        |73        |67      |
|78        |65        |82      |

```python
scores = [
    [71,62,78],
    [65,73,67],
    [78,65,82]]
scores
```

```python
scores_csv = ""
for index,student_marks in enumerate(scores):
    csv_line = ",".join(map str,student_marks)
    scores_csv += csv_line
    if index<(len(scores)-1):
        scores_csv +="\n"
print(scores_csv)        
```

```python
import csv
with open("scores.csv", "w") as f:
    f.write(scores_csv)
with open("scores.csv") as f:
    scores_csv = f.read()
print(scores_csv)
```

CSV MODULES

```python
import csv
with open('scores.csv') as f:
    #f, if the file : quoting=csv.QUOTE_NONNUMERIC, read any value that is in a string format as a number
    reader = csv.reader(f, quoting=csv.QUOTE_NONNUMERIC)
    #Conversion of an object into a list
    scores_with_csv_module = list (reader)
    #print 
scores_with_csv_module 
```

```python
#conversion of a float into an integer
scores_with_integers = [[int(value) for value in row] for row in scores_with_csv_module]
print(scores_with_integers)
```

```python
scores_with_integers = []
for row in scores_with_csv_module:
    int_row = []
    for value in row:
        int_row.append(int(value))
        scores_with_integers.append(int_row)
```

```python
import csv
with open("Employee.csv") as f:
    #read my file
    reader = csv.reader(f)
    #Convert it to a list
    employee_data = list(reader)
    
    #printing only the header and the first 5 rows of data
    for index in range(5):
        print (employee_data[index])
```

```python
type(employee_data)
```

```python
bachelors = []
#filter the rows where education is 'Bachelors'
for row in employee_data:
    if row["Education"] == "Bachelors":
        bachelors.append(row)
        #Print the number of Bachelors records
print(f"""Out of {len(employee_data)} total records, this dataset contains information about {len(bachelors)} records with a Bachelors' degree""")
```

```python
bachelors_male = []
#filter rows where 'Education' is 'Bachelors' and 'Gender' is 'Male'
for row in employee_data:
    if row ['Education'] == 'Bachelors' and row ['Gender'] == 'Male':
        #Extract only the Education and Gender fields
        bachelors_male.append({'Education': row['Education'], 'Gender': row['Gender']})
bachelors_male
```

```python
#write the filtered data to a csv data
import csv 
with open ("bachelors_male.csv", "w", newline ='') as f:
    writer = csv.DictWriter(f, fieldnames={'Education','Gender'})
    writer.writeheader()
    for rows in bachelors_male:
        writer.writerow(row)
```
