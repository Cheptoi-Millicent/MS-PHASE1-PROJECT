Handling Missing Values

```python
#Dropping the rows with missing or null values
netflix_data.dropna(subset=['director','country','cast'],how ='any',inplace=True)
netflix_data
#train_1.dropna(subset=['Age'],how='any',inplace=True)
```

```python
#By replacing  the missing values by ffill(forward fill) 
netflix_data.fillna(method = 'ffill', axis = 1)
#netflix_data['cast'] = netflix_data['cast'].ffill()
netflix_data
#book['Name'] = book['Name'].ffill()
```

```python
#By replacing  the missing values by  bfill(backwardfill)
netflix_data.fillna(method = 'bfill', axis = 1)
```
