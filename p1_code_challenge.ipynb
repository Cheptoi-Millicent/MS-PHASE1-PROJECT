### 1.2) Convert the `'Release Clause'` values from Euros to dollars

The `'Release Clause'` variable contains prices denominated in Euros. Use the exchange rate `1 Euro = 1.2 Dollars` to convert the prices to dollars. 

````python
import pandas as pd

# Sample DataFrame (Replace this with loading your actual dataset)
df = pd.read_csv("fifa.csv")

# Define the exchange rate
exchange_rate = 1.2

# Remove currency symbols and convert to numeric
# Remove the '€' symbol and '$' if it represents dollars
#df['Release Clause'] = df['Release Clause'].replace({'€': '', '$': ''}, regex=True)

# Convert to numeric, using errors='coerce' to handle any non-numeric values
df['Release Clause'] = pd.to_numeric(df['Release Clause'], errors='coerce')

# Convert Euros to Dollars
df['Release Clause'] = df['Release Clause'] * exchange_rate

# Print the DataFrame to check the results
print(df[['Release Clause']])
````

### 3.2) Create a string variable `oldest_argentine_name` and a numeric variable `oldest_argentine_age` containing the name and age (respectively) of the oldest player with Argentina nationality.

````python
#CodeGrade step3.2
#Replace None with appropriate code
import pandas as pd

# Load the dataset
df = pd.read_csv('fifa.csv')

# Filter the dataset for players with Argentina nationality
argentines = df[df['Nationality'] == 'Argentina']

# Find the oldest Argentine player
oldest_player = argentines.iloc[argentines]

# Extract the name and age of the oldest Argentine player
oldest_argentine_name  = oldest_player['Name']
oldest_argentine_age = oldest_player['Age']

# Print the results 
print(f"Oldest Argentine Player: {oldest_argentine_name}")
print(f"Age: {oldest_argentine_age}")
````

### 4.3) Define a function `get_players_on_team()` that returns a list of names of all the players who have played on a given team.

Your function should take two arguments:

- A dictionary of player information
- A string containing a team name (for which you are trying to find the player names)

````python
# CodeGrade step4.3

def get_players_on_team(players, team_name):
    player_names = []
    # Your code here
    
    # Iterate through the dictionary of players
    for player_name, player_info in players.items():
        # Check if the team_name is in the player's list of teams
        if team_name in player_info.get('teams', []):
            # Append the player's name to the list
            player_names.append(player_name)
    
    return player_names
````

````python
# This test confirms that get_players_on_team() returns the right names for Manchester United

manchester_united_players = ['Cristiano Ronaldo', 'De Gea']
players_on_manchester_united = get_players_on_team(players, 'Manchester United')

assert players_on_manchester_united == manchester_united_players
````
